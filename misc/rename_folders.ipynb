```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rename_legend = pd.read_csv("/Volumes/DATA/magda_figure_images_rename/File_Name_Changes_CD_Provissionally_accepted.csv")
```

```python
# get folder names

rename_legend["OLD_folder"] = rename_legend["OLD"].apply(lambda x: x.split("/")[6])
rename_legend["NEW_folder"] = rename_legend["NEW"].apply(lambda x: x.split("/")[6])
```

```python
rename_legend
```

```python
import subprocess
in_folder = "/Volumes/DATA/magda_figure_images_rename/Figure_images/"
out_folder = "/Volumes/DATA/magda_figure_images_rename/Figure_images_renamed/"

for index, row in rename_legend.iterrows():
    old_folder = row["OLD_folder"]
    new_folder = row["NEW_folder"]
    command = f"mv {in_folder}{old_folder} {out_folder}{new_folder}"
    print(command)
    subprocess.run(command, shell=True)

```
